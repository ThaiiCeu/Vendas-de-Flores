```python
# Catálogo_flores
flores = [
    {
        "nome": "Rosas",
        "preco": 10.00,
        "descricao": "Disponível em várias cores, ideal para presentes românticos e arranjos."
    },
    {
        "nome": "Margaridas",
        "preco": 5.00,
        "descricao": "Simples e delicadas, ótimas para buquês variados."
    },
    {
        "nome": "Orquídeas",
        "preco": 25.00,
        "descricao": "Flores elegantes e sofisticadas, com longa duração e beleza única."
    },
    {
        "nome": "Girassóis",
        "preco": 8.00,
        "descricao": "Grandes e vibrantes, perfeitos para alegrar qualquer ambiente."
    },
    {
        "nome": "Tulipas",
        "preco": 12.00,
        "descricao": "Disponíveis em várias cores, simbolizam amor e carinho."
    },
    {
        "nome": "Lírios",
        "preco": 15.00,
        "descricao": "Flores perfumadas e elegantes, ideais para decoração e presentes especiais."
    }
]

# Carrinho_ompras
carrinho = []

# Catálogo
def exibir_catalogo():
    print("\nCatálogo de Flores:")
    i = 0
    while i < len(flores):
        flor = flores[i]
        print(f"{i + 1} - {flor['nome']}: R${flor['preco']:.2f} - {flor['descricao']}")
        i = i + 1

# Adicionar_carrinho
def adicionar_ao_carrinho():
    exibir_catalogo()
    escolha = int(input("\nDigite o número da flor que deseja adicionar ao carrinho: ")) - 1
    if escolha >= 0 and escolha < len(flores):
        quantidade = int(input(f"Quantas {flores[escolha]['nome']} você deseja adicionar? "))
        i = 0
        encontrado = False
        while i < len(carrinho):
            if carrinho[i]["nome"] == flores[escolha]["nome"]:
                carrinho[i]["quantidade"] = carrinho[i]["quantidade"] + quantidade
                encontrado = True
            i = i + 1
        if not encontrado:
            carrinho.append({"nome": flores[escolha]["nome"], "quantidade": quantidade, "preco": flores[escolha]["preco"]})
        print(f"{quantidade} {flores[escolha]['nome']} adicionadas ao carrinho.")
    else:
        print("Opção inválida.")

# Exibir_carrinho
def exibir_carrinho():
    if len(carrinho) == 0:
        print("\nSeu carrinho está vazio.")
    else:
        print("\nSeu Carrinho:")
        total = 0
        i = 0
        while i < len(carrinho):
            item = carrinho[i]
            subtotal = item["quantidade"] * item["preco"]
            total = total + subtotal
            print(f"{item['nome']}: {item['quantidade']} unidade(s) - R${subtotal:.2f}")
            i = i + 1
        print(f"Total: R${total:.2f}")

# Finalizar_compra
def finalizar_compra():
    if len(carrinho) == 0:
        print("\nSeu carrinho está vazio. Adicione flores antes de finalizar.")
    else:
        exibir_carrinho()
        confirmacao = input("\nDeseja finalizar a compra? (sim/não): ").lower()
        if confirmacao == "sim":
            print("\nCompra finalizada! Agradecemos por comprar conosco! (^-^)/")
            while len(carrinho) > 0:
                carrinho.pop()
        else:
            print("\nCompra não finalizada.")

# Menu_principal
while True:
    print("\nMenu:")
    print("1 - Exibir catálogo")
    print("2 - Adicionar ao carrinho")
    print("3 - Exibir carrinho")
    print("4 - Finalizar compra")
    print("5 - Sair")
    opcao = input("Escolha uma opção: ")

    if opcao == "1":
        exibir_catalogo()
    elif opcao == "2":
        adicionar_ao_carrinho()
    elif opcao == "3":
        exibir_carrinho()
    elif opcao == "4":
        finalizar_compra()
    elif opcao == "5":
        print("Agradecemos por comprar na nossa loja. Até! (^-^)/")
        break
    else:
        print("Opção não encontrada. Digite novamente.")
```
